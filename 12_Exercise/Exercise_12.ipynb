In this notebook we are going to use two different neural network, namely a Deep neural Network (DNN) and a Convolutional Neural Network (CNN), to recognize the 10 digits handwritten in a image with 28x28 pixels 

```python
import tensorflow as tf
from tensorflow import keras
import os
import numpy as np
import matplotlib.pyplot as plt
import time
from PIL import Image

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import SGD, Adam, RMSprop, Adagrad, Adadelta, Adam, Adamax, Nadam
from tensorflow.keras.layers import Flatten, Conv2D, MaxPooling2D


seed=int(time.time())
np.random.seed(seed) # fix random seed
tf.random.set_seed(seed)
```

As a first thing  we import the dataset and process the data

```python
# input image dimensions
img_rows, img_cols = 28, 28 # number of pixels 
# output
num_classes = 10 # 10 digits

# the data, split between train and test sets
(X_train, Y_train), (X_test, Y_test) = tf.keras.datasets.mnist.load_data()

print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
print()
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
```

```python
# reshape data, it could depend on Keras backend
X_train = X_train.reshape(X_train.shape[0], img_rows*img_cols)
X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)

# cast floats to single precesion
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# rescale data in interval [0,1]
X_train /= 255
X_test /= 255

# look at an example of data point
print('an example of a data point with label', Y_train[27])
# matshow: display a matrix in a new figure window
plt.matshow(X_train[27,:].reshape(28,28),cmap='binary')
plt.show()


# convert class vectors to binary class matrices, e.g. for use with categorical_crossentropy
Y_train = keras.utils.to_categorical(Y_train, num_classes)
Y_test = keras.utils.to_categorical(Y_test, num_classes)

print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
```

### EXERCISE 12.1 

Now that we have prepared the dataset, we can create a feed forward DNN to recognize the data. This network will have input and output neurons fixed respectively to 784 and 10, but the neurons in the hidden layers and the number of the layers itself may vary from structure to structure.

In this exercise we are going to create different networks, all with the same structure but with different otpimizers, in order to see which performs the best on this problem.

- **Create**

```python
def create_DNN():
    # instantiate model
    model = Sequential()
    # add a dense all-to-all relu layer
    model.add(Dense(400,input_shape=(img_rows*img_cols,), activation='relu'))
    # add a dense all-to-all relu layer
    model.add(Dense(100, activation='relu'))
    # apply dropout with rate 0.5
    model.add(Dropout(0.5))
    # soft-max layer
    model.add(Dense(num_classes, activation='softmax'))
    
    return model
```

- **Compile**

```python
def compile_model(optim):
    # create the model
    model=create_DNN()
    # compile the model
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer=optim,
                  metrics=['accuracy'])
    return model
```

- **Train different models**

```python
# training parameters
batch_size = 30
epochs = 10

models=[]
histories=[]

# Create networks
models.append(compile_model('sgd'))
models.append(compile_model('Adam'))
models.append(compile_model('Adagrad'))
models.append(compile_model('Adamax'))

count=1

for model in models:
    print("Training model: ",count)
    history = model.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=False,
          validation_data=(X_test, Y_test))
    print("Training successful!")
    
    histories.append(history)
    count+=1
```

- **Evaluate the networks on unseen data**

```python
labels=['sgd','Adam','Adagrad','Adamax']
# evaluate model
scores =[]

for model in models:
    scores.append(model.evaluate(X_test, Y_test, verbose=1))

# print performance
for j in range(len(models)):
    print("===================")
    print(labels[j])
    print('Test loss:', scores[j][0])
    print('Test accuracy:', scores[j][1])

print("===================")
```

- **Plot the training histories**

```python
# look into training history

fig, axs=plt.subplots(2,2, figsize=(12,12))

fig.suptitle("Comparison between diferent optimizers on same structure")

# summarize history for accuracy

for i in range(len(models)):
    axs[0][0].plot(histories[i].history['accuracy'],label=labels[i])
    axs[0][1].plot(histories[i].history['val_accuracy'],label=labels[i])
    axs[1][0].plot(histories[i].history['loss'],label=labels[i])
    axs[1][1].plot(histories[i].history['val_loss'],label=labels[i])
    

axs[0][0].set_ylabel("Accuracy")
axs[0][0].set_xlabel("Epoch")
axs[0][0].set_title("Accuracy comparison")

axs[0][1].set_ylabel("Validation accuracy")
axs[0][1].set_xlabel("Epoch")
axs[0][1].set_title("Accuracy comparison on validation set")

axs[1][0].set_ylabel("Loss")
axs[1][0].set_xlabel("Epoch")
axs[1][0].set_title("Loss comparison")

axs[1][1].set_ylabel("Validation_loss")
axs[1][1].set_xlabel("Epoch")
axs[1][1].set_title("Loss comparison on validation set")

for i in range(2):
    for j in range(2):
        axs[i][j].grid(True)
        axs[i][j].legend(loc='best')


plt.show()
plt.close()
```

We are using the same structure so the only difference between the networks is the optimizers, but also the initialization of the weights and biases, in fact not all the curves begin at the same point.

- **Use the trained network to predict hadwritten digits**

```python
for i in range(len(models)):
    print("====================")
    print(labels[i])
    print("====================")
    X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)
    predictions = models[i].predict(X_test)

    X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols,1)

    plt.figure(figsize=(15, 15)) 
    for i in range(10):    
        ax = plt.subplot(2, 10, i + 1)    
        plt.imshow(X_test[i, :, :, 0], cmap='gray')    
        plt.title("Digit: {}\nPredicted:    {}".format(np.argmax(Y_test[i]), np.argmax(predictions[i])))    
        plt.axis('off') 
    plt.show()
```

### EXERCISE 12.2

Now we will consider a different type of feed forward network, a convolutional neural network. This type of network isn't fully connected and has different type of layers (convolutional, pooling, dense) in order to replicate the human retina mechanism to recognize shape and edges.

In order to use a CNN on the MNIST dataset, the first thing to do is to reshape it.

```python
# reshape data, depending on Keras backend
if keras.backend.image_data_format() == 'channels_first':
    X_train = X_train.reshape(X_train.shape[0], 1, img_rows, img_cols)
    X_test = X_test.reshape(X_test.shape[0], 1, img_rows, img_cols)
    input_shape = (1, img_rows, img_cols)
else:
    X_train = X_train.reshape(X_train.shape[0], img_rows, img_cols, 1)
    X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)
    input_shape = (img_rows, img_cols, 1)
    
print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
print()
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
```

- **Create and initialize the network**

By adding Conv2D, Pooling and Dense layers we are going to create the network

```python
def create_CNN():
    # instantiate model
    model = Sequential()
    
    # add first convolutional layer 
    model.add(Conv2D(40, kernel_size=(5, 5),
                     activation='relu',
                     padding='same',
                     input_shape=(28,28,1)))
    # Pooling layer 14x14x1
    model.add(MaxPooling2D(pool_size=(2, 2)))
    # Second convolutional layer
    model.add(Conv2D(10, kernel_size=(5, 5),
                     activation='relu'))
    
    
    #Flatten layer
    model.add(Flatten())

    #fully connected layer
    model.add(Dense(100,activation='tanh'))
    #Output layer
    model.add(Dense(10, activation='softmax'))
    
    # compile the model
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer='SGD',
                  metrics=['accuracy'])
    return model
```

```python
my_cnn=create_CNN()

my_cnn.summary()
```

- **Train the network and evaluate performances**

```python
# training parameters
batch_size = 30
epochs = 10

# train CNN
history =my_cnn.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test, Y_test))

# evaliate model
score = my_cnn.evaluate(X_test, Y_test, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

- **Plot the training history**

```python
# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.ylabel('model accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='best')
plt.show()

# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.ylabel('model loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='best')
plt.show()
```

With only 10 epochs of learning we reach almost 99% of accuracy on the training set. Loss in the training set decreases faster than on the test set, so we reach a point where the training error is lower than the test set, but this is ok since we aren't overfitting on the training set yet, the test error is decreasing.


- **Prediction**

```python
#X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)
predictions = my_cnn.predict(X_test)

X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols,1)

print("Convolutional neural network")

plt.figure(figsize=(15, 15)) 
for i in range(10):    
    ax = plt.subplot(2, 10, i + 1)    
    plt.imshow(X_test[i, :, :, 0], cmap='gray')    
    plt.title("Digit: {}\nPredicted:  {}".format(np.argmax(Y_test[i]), np.argmax(predictions[i])))    
    plt.axis('off') 
plt.show()
```
