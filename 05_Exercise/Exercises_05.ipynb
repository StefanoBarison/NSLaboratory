In this notebook we will use the Metropolis algorithm in order to obtain an estimation of the expectation value of the radius of $H$ when the electron has a wavefunction $\Psi_{1,0,0}$  or $\Psi_{2,1,0}$.

Since the evaluation of an expectation value is the calculation of an integral and the squared modulus $|\Psi(\vec{r},t)|^2$ can be interpreted as a density probability function, the use of the Metropolis algorithm is well suited.

### EXERCISE 05.1

As said before, in this exercise the calculation of $\left\langle r \right\rangle_{\Psi_{1,0,0}}$ and $\left\langle r \right\rangle_{\Psi_{2,1,0}}$ is required.

First things first, we expand the expression in order to see the integral we have to calculate:

$$ \left\langle r \right\rangle_{\Psi_{1,0,0}}= \int  d\vec{r} \quad r|\Psi(\vec{r})|^2 = \int dx dy dz \quad \sqrt{x^{2}+y^{2}+z^{2}}\frac{a_{0}^{-3}}{\pi}e^{-\frac{2\sqrt{x^{2}+y^{2}+z^{2}}}{a_{0}}} $$


Since $\int d\vec{r} |\Psi(\vec{r})|^2 = 1$, this will be our probability density function from which we have to extract the sample. The sampling procedure will be done by using the Metropolis algorithm.

1) As the first thing, we show the calculation of the expectation value of $r$ in $a_0$ units ($a_{0}=1$) using $|\Psi_{1,0,0}|^{2}$ as probability density function. The evaluation of the integral and of its related uncertainty is done by the usual blocking tecnique

```python
import numpy as np
import math
import matplotlib 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
data1=open("1s_integral.dat",'r').read().split('\n')

data1=[x.split(',') for x in data1]

x1=np.zeros(len(data1)-1)
y1=np.zeros(len(data1)-1)
y1_err=np.zeros(len(data1)-1)



for i in range(len(data1)-1):
    x1[i]=data1[i][0]
    y1[i]=data1[i][1]
    y1_err[i]=data1[i][2]


plt.errorbar(x1,y1-1.5,yerr=y1_err,color='r',lw=2,ecolor='gray')
plt.xlabel('Block')
plt.ylabel('<r>-3/2 in $a_0$ units')
plt.grid(True)
plt.title("Expected value of r for $\Psi_{1,0,0}$")

plt.show()

plt.close()
```

We can also see the points that we have generated with the Metropolis sampling technique directly in 3D

```python
data2=open("1s_points.dat",'r').read().split('\n')
data2=[x.split(',') for x in data2]

x=np.zeros(len(data2)-1)
y=np.zeros(len(data2)-1)
z=np.zeros(len(data2)-1)

for j in range(len(data2)-1):
    x[j]=data2[j][0]
    y[j]=data2[j][1]
    z[j]=data2[j][2]
    
ax = Axes3D(plt.figure())
ax.scatter(x, y, z, c=z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
ax.set_title("Points sampled using $|\Psi_{1,0,0}|^{2}$ as probability distribution")

plt.show()
plt.close()
```

In the end, I would like to answer all the questions related to the Metropolis algorithm and its initialization:

1) In both cases, I used the origin as the starting point of the algorithm. This choice is motivated by the fact that for the 1s orbital we consider the most probable space if we start from the origin and move in the neighborhood, while for the 2p we avoid being captured by one of the two lobes. I chose also a different starting point , namely (10,10,10) for both, and saw that the integral isslower to converge

```python
data3=open("1s_integral_far.dat",'r').read().split('\n')
data3=[x.split(',') for x in data3]

data4=open("2p_integral_far.dat",'r').read().split('\n')
data4=[x.split(',') for x in data4]

x3=np.zeros(len(data3)-1)
y3=np.zeros(len(data3)-1)
y3_err=np.zeros(len(data3)-1)

x4=np.zeros(len(data4)-1)
y4=np.zeros(len(data4)-1)
y4_err=np.zeros(len(data4)-1)

for i in range(len(data3)-1):
    x3[i]=data3[i][0]
    y3[i]=data3[i][1]
    y3_err[i]=data3[i][2]
    
for i in range(len(data4)-1):
    x4[i]=data4[i][0]
    y4[i]=data4[i][1]
    y4_err[i]=data4[i][2]



fig, (ax1,ax2)=plt.subplots(1,2,figsize=(13,6),dpi=80)

ax1.errorbar(x3,y3-1.5,yerr=y3_err,color='g',lw=2,ecolor='gray')
ax1.set(xlabel='block',ylabel='<r>-3/2 in $a_0$ units')
ax1.grid(True)
ax2.errorbar(x4,y4-5,yerr=y4_err,color='b',lw=2,ecolor='gray')
ax2.set(xlabel='block',ylabel='<r>-5 in $a_0$ units')
ax2.grid(True)

fig.suptitle("Evaluation of the integral with starting point (10,10,10)")

plt.show()

plt.close()

```
